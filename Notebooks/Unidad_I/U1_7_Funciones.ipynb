# <font color="red">1.7 Funciones en Python</font>

Las funciones son estructuras que permiten definir un bloque de código reutilizable, es decir, que se puede ejecutar muchas veces dentro del programa. Si sabemos manejarlos correctamente pueden simplificarnos muchísimo la vida.

#### **Ventajas de las funciones**:

*   Divider y ordenar el código en partes más sencillas para depurar y programar con mayor facilidad.
*   Reutilizar el código, evitando así repeticiones innecesarias dentro de un programa.


Este último ayuda a cumplir con un principio de desarrollo de software que se llama DRY (Don't Repeat Yourself).

#### **Sintaxis básica**
Las funciones en Python reciben datos de entrada llamados argumentos, los cuales indica el usuario, posteriormente los procesa y finalmente devuelve
datos de salida. La cantidad y tipo de parametros depende escencialmente de lo que busquen resolver.

#### Orden de los argumentos

Cuando hay varios argumentos en una función, estos se pueden agregar de dos formas:

* **Posicional**: Los argumentos se agregan en el mismo orden en que aparecen los parámetros correspondientes en la definición de la función.

* **Nominal**: Sin importar el orden, se especifica el nombre del parámetro al que se asocia un argumento.

```python
#Ejemplo: Invocación posicional (el orden importa)

def anio_nacimiento(anio_actual,edad):
  print(f"Año de nacimiento: {anio_actual-edad}, edad: {edad}")
    
anio_nacimiento(anio_actual=2026,edad=25) 
```

```python
anio_nacimiento(2026,25) # Ambas maneras de invocar la función dan el mismo resultado
```

```python
#Ejemplo: Invocación nominal
#Usando la misma función
anio_nacimiento(anio_actual=2026,edad=25)
anio_nacimiento(edad=25,anio_actual=2026) #Mientras el nombre del argumento sea correcto, no existirá problema alguno
```

```python
# OJO: No pueden combinar una invocación nominal antes de la posición
anio_nacimiento(edad = 25, 2026)  #Error
```

#### Las funciones no solo regresan tuplas, también pueden regresar diferentes estructuras como listas o diccionarios.

#### <font color="orange"> Ejemplo: Regresar una lista

```python
def regresar_lista(elemento_1, elemento_2, elemento_3, elemento_4):
    return [elemento_1, elemento_2, elemento_3, elemento_4]
```

```python
# Almacenar e imprimir la lista de tipos de paisajes kársticos
d_list = regresar_lista('caverna', 'dolina', 'cañón', 'estalactitas')
print(d_list)
```

#### <font color="orange"> Ejemplo: Regresar un diccionario


```python
def regresar_dict(key_1, key_2, value_1, value_2):
    d = dict()  #Creamos un diccionario vacío
    d[key_1] = value_1  #Agregando elementos al diccionario
    d[key_2] = value_2
    return d
```

```python
#Tipos de intemperismo
d_dict = regresar_dict('Físico', 'Biológico', '', '')
print(d_dict)
```

# Funciones con argumentos variables: ***args** y  ****kwargs**

A veces cuando trabajamos con funciones es posible que el número de parámetros sea variable. `*args` y `**kwargs` son herramientas que permiten a las funciones aceptar un número variable de argumentos sin la necesidad de definirlos uno por uno (<https://python-intermedio.readthedocs.io/es/latest/args_and_kwargs.html>). Esto las hace más flexibles y adaptables a diferentes escenarios.


## `*args` (Non-Keyword Arguments)
El nombre de args viene de argumentos, que es como se denominan en programación a los parámetros de entrada de una función. En las funciones, la sintáxis de `*args` se usa para pasar una variable que sea "non-keyworded" (o sea no nominal) y con tamaño variable. Por convención, para usarla se coloca un asterisco seguido de la palabra args (`*args`), pero en realidad puedes usar cualquier palabra en vez de *args*.

#### <font color="orange"> Ejemplo 1:

```python
def estudiantes(*args):
    print(args)  #Los valores que coloquen dentro del argumento args serán de tipo tupla
    
estudiantes('Montse', 'Adriana', 'Eros', 'Rodrigo', 'Jorge', 'América', 'Julia') #Todo este conjunto de strings son un solo argumento, sin necesidad de hacerlo lista
```

```python
# Imprimir varios elementos
def estudiantes_1(*args):  #args es el nombre de tu argumento
    for i in args:
        print(i)
        
estudiantes_1('Montse', 'Adriana', 'Eros', 'Rodrigo', 'Jorge', 'América', 'Julia') #Todo este conjunto de strings son un solo argumento, sin necesidad de hacerlo lista
```

```python
def estudiantes_2(*nombres):
    for i in nombres:
        print(i)

estudiantes_2('Montse', 'Adriana', 'Eros', 'Rodrigo', 'Jorge', 'América', 'Julia')
```

```python
# Pero recuerden, no siempre hay una única solución. Por ejemplo en vez de usar *args, podríamos usar una variable pero que esa variable sea una lista
def estudiantes_3(names):
    for i in names:
        print(i)

estudiantes_3(['Montse', 'Adriana', 'Eros', 'Rodrigo', 'Jorge', 'América', 'Julia']) 
```


##   `**kwargs` (Keyword Arguments)

`**kwargs` permite pasar argumentos de longitud variable asociados con un nombre o key a una función,  por lo que se puede pensar en *kwargs* como si fuera una especie de diccionario. Para usarla se coloca doble asterisco seguido de la palabra kwargs (`*kwargs`), pero en también puedes usar cualquier palabra en vez de *kwargs*.


#### <font color="orange"> Ejemplo 1:

```python
def mostrar_datos(**kwargs):
    print(kwargs)
```

```python
#Forma de llamar la función
mostrar_datos(nombre="Teo", materia="HCA", calificacion=10, ciclo="ordinario") #OJO: Para este caso, deben usar "=" cuando definan el valor del key 
                                                                                #Puedes ingresar n keys con sus respectivos values
                                                                              
```

```python
#Iterando
def mostrar_datos1(**kwargs):
    for i in kwargs: 
        print(i)
```

```python
#¿Qué me regresa esta función?
mostrar_datos1(nombre="Teo", materia="HCA", calificacion=10, ciclo="ordinario")
```
