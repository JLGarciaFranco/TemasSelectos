## <font color="darkblue"> Gráficos de pastel

Un gráfico de pastel es un elemento circular que muestra cómo se divide un total en proporciones o porcentajes. Normalmente los gráficos de pastel se utilizan en negocios para representar operaciones como ventas, resultados de cuestionarios, etc. Son de gran utilidad porque dan un resumen rápido.
En Matplotlib se pueden hacer gráficos de pastel utilizando la función `.pie()`.

#### <font color="orange"> Ejemplo 1: Mascotas en una veterinaria

```python
# Vamos a inventar algunos datos sobre mascotas

especies= ['Michis', 'Lomitos', 'Peces Beta', 'Cuyos', 'Tortugas'] #Etiquetas
cantidad = [30, 25, 10, 3,8 ]  #Cantidad de mascotas

#Gráfica de pastel, usando .pie()
plt.figure(figsize=(5,5))

plt.pie(cantidad,labels=especies)

plt.show()
```

**Noten algo**, cada una de las partes del plot tiene una cuña que indica el valor/proporción de cada uno de los elementos en relación al total. Por default, la primer marca empieza desde el eje x y se mueve en contra de las manecillas del reloj.

#### <font color="orange"> Ejemplo 2: Cambiemos el ángulo en el que empieza el primer elemento `startangle`

```python
# Por default empieza en 0°
plt.figure(figsize=(5,5))

plt.pie(cantidad,labels=especies,startangle = 45)

plt.show()
```

#### <font color="orange"> Ejemplo 3: Resaltando un valor `explode`

```python
#Definimos los valores que indicarán qué tan lejos del centro estarán las cuñas
exp = [0.2, 0.0, 0.0, 0.0, 0.0] # El orden es el que se va graficando, por ejemplo el primero es el de los gatos y el segundo de los lomitos y así sucesivamente

plt.figure(figsize=(5,5))

plt.pie(cantidad,labels=especies,startangle = 45, explode=exp)

plt.show()
```

#### <font color="orange"> Ejemplo 4: Efecto de sombra `shadow`

```python
exp = [0.2, 0.0, 0.0, 0.0, 0.0]

plt.figure(figsize=(5,5))

plt.pie(cantidad,labels=especies,startangle = 45, explode=exp,shadow=True)

plt.show()
```

### <font color="brown">Resumen de parámetros útiles para mejorar nuestros gráficos de pastel
* `startangle`: Rota el gráfico (por default 0$^\circ$)
* `explode`: Separa una cuña del centro (destacar categoría).
* `shadow`: Agrega sombra.
* `colors`: Nos permite definir los colores manualmente.
* `autopct`: Muestra el porcentaje automáticamente.
* `labeldistance`: Controla qué tan lejos aparecen los labels.
* `pctdistance`: Controla la posición del porcentaje.
* `wedgeprops`: Controla propiedades de las cuñas.

#### <font color="orange"> Ejemplo 8: Pasando de gráfico de pastel a gráfico de dona 

```python
import matplotlib.pyplot as plt

valores = [40, 35, 25]
labels = ["Marítimo", "Continental", "Montaña"]

plt.pie(
    valores,
    labels=labels,
    autopct="%1.1f%%",
    startangle=90,
    pctdistance=0.8,
    wedgeprops={"width": 0.5, "edgecolor": "black"}) #'width' crea el hueco y controla qué tan grueso es el anillo
                                                    #width = 1 → pastel normal
                                                    #width < 1 → dona



plt.title("Tipos de masas de aire",color="red")

plt.show()
```

#### <font color="orange">Ejemplo 9: Cambiando los colores de los labels y los porcentajes

```python
import matplotlib.pyplot as plt

valores = [40, 35, 25]
labels = ["Marítimo", "Continental", "Montaña"]

plt.pie(
    valores,
    labels=labels,
    autopct="%1.1f%%",
    startangle=90,
    pctdistance=0.8,
    wedgeprops={"width": 0.5, "edgecolor": "black"},
    textprops={"color": "darkblue"})  #textprops nos permite cambiar formato del texto como el color



plt.title("Tipos de masas de aire",color="red")

plt.show()
```

**También podemos definir un color diferente para cada label**
(<https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html>).

```python
import matplotlib.pyplot as plt

valores = [40, 35, 25]
labels = ["Marítimo", "Continental", "Montaña"]
color_labels=["blue","orange","green"]

_,texts,_=plt.pie(
    valores,
    labels=labels,
    autopct="%1.1f%%",
    startangle=90,
    pctdistance=0.8,
    wedgeprops={"width": 0.5, "edgecolor": "black"}) 

#Cambiando el color de los labels
for i, text in enumerate(texts):
    text.set_color(color_labels[i]) #'set_color' es una función que permite cambiar el color de un objeto gráfico o visual.
    

plt.title("Tipos de masas de aire",color="red")

plt.show()
```

## <font color="darkblue"> Gráficos de barras

Un gráfico de barras se usa para **comparar cantidades entre categorías**. Es ideal cuando tienes datos discretos, es decir, aquellos que se puedan contar. En Matplotlib ocupamos la función `.bar` para gráficos de barras.

### <font color="orange"> Ejemplo 1: Alumnos por orientación

```python
#Definimos nuestros datos
estudiantes = ["Sólidos", "Atmosféricos", "Espaciales", "Acuáticos", "Ambientales"]
cantidad = np.array([3, 8, 1, 10, 3])

#Graficamos
plt.bar(estudiantes, #Categorías o posiciones
        cantidad, #Altura (valores por categoría)
        color = ['#bc6c25' , '#3a86ff', '#14213d', '#0081a7', '#a7c957'], #Colores para cada barra
        width = 0.7)#Ancho de cada barra

plt.title('Cantidad de alumnos por orientación ')
plt.grid(linestyle="--", color="gray",alpha=0.2)
plt.xlabel('Orientación')
plt.ylabel('Alumnos')
plt.show()
```

```python
# Podemos cambiar la orientación de las barras, ocupando `plt.barh`

#Graficando
plt.barh(estudiantes, 
         cantidad,
         color = ['#bc6c25' , '#3a86ff', '#14213d', '#0081a7', '#a7c957'], 
         height = 0.8) # altura de las barras (valores en eje y), en pocas palabras el grosor en barras horizontales

plt.title('Cantidad de alumnos por orientación ')
plt.xlabel('Orientación')
plt.grid(linestyle="--", color="gray",alpha=0.2)
plt.ylabel('Alumnos')
plt.show()
```

#### <font color="orange"> Ejemplo 2: Múltiples barras por categoría

```python
#Vamos a mostrar características de diferentes especies de pingüinos (Los datos vinieron de aqui https://allisonhorst.github.io/palmerpenguins/)

#Especies de pingüinos
especie = ["Adelie", "Chinstrap", "Gentoo"]

adelite  = [18.35, 38.79 ,189.95]  #Cada lista contiene el ancho de su pico, largo del pico y largo de la la aleta
chinstrap = [18.43, 48.83, 195.82]
gentoo = [14.98, 47.50, 217.19]

#Defino el ancho de cada barra
width = 0.20

#GRAFICANDO

#Defino cuántas barras tendré por categoría
x = np.arange(3) #Asignamos un índice a cada categoría

plt.bar(x-0.2, adelite, width,color ='#d9f0ff')
plt.bar(x, chinstrap, width, color ='#a3d5ff')
plt.bar(x+0.2, gentoo, width, color ='#83c9f4')

#Definimos los nombres de las categorías de cada eje
plt.xticks(x,['Ancho Pico', 'Largo Pico', 'Largo Aleta'] )

plt.xlabel("Característica")
plt.grid(linestyle="--", color="gray",alpha=0.2)
plt.ylabel("Largo en mm")
plt.legend(especie)
plt.show()
```

**Con los mismos datos podemos hacer una representación diferente, lo importante es que sepan qué están haciendo.**

```python
#Ahora quiero que cada categoría sea la especie y las barras sean sus características
especie = ["Adelie", "Chinstrap", "Gentoo"]


#Carácterísticas
ancho_p = [18.35, 18.43, 14.98] #Adelie, Chinstrap, Gentoo
largo_p = [38.79, 48.83, 47.50] 
largo_aleta = [189.95, 195.82, 217.19]

width = 0.20

#GRAFICANDO
x = np.arange(3) 
plt.bar(x-0.2, ancho_p, width,color ='#d9f0ff',edgecolor="black") #'edcolor' color a los bordes
plt.bar(x, largo_p, width, color ='#a3d5ff',edgecolor="black")
plt.bar(x+0.2, largo_aleta, width, color ='#83c9f4',edgecolor="black")
plt.xticks(x,especie )
plt.xlabel("Característica")
plt.ylabel("Largo en mm")
plt.grid(linestyle="--", color="gray",alpha=0.2)
plt.legend(['Ancho Pico', 'Largo Pico', 'Largo Aleta'], loc = 'upper left', ncols = 3) #'ncols' nos permite definir la cantidad de columnas 
plt.show()                                                                             #dentro de una leyenda
```

#### <font color="orange"> Ejemplo 3: Barras apiladas (stacked bar)

```python
#Queremos ver los días con lluvias/sin lluvias y frentes fríos

categorias = ["México","España"]

heat_waves = [50,55]
cold_waves = [20,10]

plt.bar(categorias, heat_waves,width=0.5,label="Olas de calor",color="brown")
plt.bar(categorias, cold_waves,width=0.5,bottom=heat_waves, label="Olas de frío",color="darkblue")  #'bottom' permite apilar

plt.grid(linestyle="--", color="gray",alpha=0.2)
plt.legend(loc="upper center")
plt.show()
```
