# Cálculo del NDVI


El índice de vegetación de diferencia normalizada (NDVI) utiliza la relación entre la luz infrarroja cercana y la luz roja dentro del espectro electromagnético.

Para calcular el NDVI, se utiliza la siguiente fórmula:

> (NIR - Rojo) / (NIR + Rojo)


NIR representa la luz infrarroja cercana y el rojo representa la luz roja.

Para los datos ráster, se toma el valor de reflectancia en las bandas roja e infrarroja cercana para calcular el índice. Este nos ayuda a saber el estado de la salud de las plantas.

Primero:
- Descarga los datos de https://ndownloader.figshare.com/files/10960109

- A continuación, abra los datos del NAIP con los que desea calcular el NDVI. Utilizará los datos de 2015:
```
datos/cold-springs-fire/naip/m_3910505_nw_13_1_20150919/crop/m_3910505_nw_13_1_20150919_crop.tif
```
- Calcule el NDVI con una matriz numpy. En este caso, las bandas que resta provienen del mismo archivo de datos. El formato de archivo tif requiere que todas las capas estén en el mismo CRS y tengan el mismo tamaño, por lo que se asume que los datos están alineados. Por lo tanto, no es necesario comprobar si los datos tienen la misma forma, CRS y extensión.

- Para calcular el NDVI, obtenga la diferencia normalizada de las capas 0 y 3.

- Obtenga la distribución de valores de NDVI.
