## <font color="darkblue"> Gráficos de línea
Son representaciones donde los puntos de datos se conectan mediante líneas para mostrar cómo una variable cambia respecto a otra. El gráfico de línea es ideal para seguir la evolución de una variable en el tiempo o frente a otra magnitud. Para este tipo de gráficos usamos la función `plt.plot`.

#### <font color="orange"> Ejemplo 1:

````python
# Para hacer gráficos de línea es necesario que definan un valor para el eje x (independiente) y otro para el eje y (dependiente)

#Definimos los valores de x y y
x=np.linspace(0,10,25)
y= np.linspace(0,5,25)

#Para gráficar llamamos a la función plot
plt.plot(x,y)
````

#### <font color="darkblue">**Además de las funciones para enchular la gráfica, existen parámetros para embellecer las líneas. Los parámetros más comúnes son: el estilo de línea, su grosor, color y agregar marcadores.**

Ejemplos de parámetros en gráficos de línea:

- **Linewidth** (`linewidth`): Este parámetro controla el grosor de la línea en el gráfico. Puedes especificar el grosor de la línea utilizando un valor numérico. Por ejemplo, linewidth=2 establecerá el grosor de la línea en 2 puntos. 

- **Color** (`color` o `c`): Este parámetro controla el color de la línea en el gráfico. Puedes especificar el color utilizando nombres de colores estándar de matplotlib (como 'blue', 'red', 'green', etc.), códigos hexadecimales ('#FF5733'), o abreviaturas de color ('b' para azul, 'r' para rojo, 'g' para verde, etc.) Para ver todos los colores, puedes consultar: <https://matplotlib.org/stable/gallery/color/named_colors.html>

- **Linestyle** (`linestyle`): Este parámetro controla el estilo de la línea en el gráfico. Puedes especificar diferentes estilos de línea, como sólida ('-'), discontinua ('--'), punteada (':'), o guiones y puntos ('-.')(<https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D.set_linestyle>).

- **Markers** (`marker`): Este parámetro permite agregar marcadores al gráfico de líneas. Estos nos ayudan a indicar la posición exacta de cada dato sobre la línea. Existen diferentes marcadores, como puntos ('.') , círculos ('o'), cuadrados ('s'), etc. En la siguiente liga podrán encontrar los markers posibles: <https://matplotlib.org/stable/api/markers_api.html>. Para modificar el tamaño del marker se usa el parámetro `markersize`.

- **Transparencia** (`alpha`): Este parámetro controla la transparencia de los elementos del gráfico. Los valores van de 0 a 1, donde un `alpha=0` el gráfico es completamente transparente y un `alpha=1`, completamente opaco.

 
**Tambíen, podemos modificar más detalles de la gráfica como el tamaño de los ejes, de las etiquetas y de los títulos.** Dentro de `plt.xlabel`, `plt.ylabel` y `plt.title` podemos modificar el tamaño de la letra con la palabra (key) `fontsize` y luego un número que es el número de puntos de la tipografía.

Además, **podemos modificar la leyenda con diferentes kwargs** como `shadow`. Para ver todos los cambios posibles, ver: <https://matplotlib.org/stable/api/legend_api.html>

#### <font color="orange"> Ejemplo 4: Styleando

### <font color="purple"> Ejercicio 1: Funciones trigonométricas

Evaluén las funciones de $sen(x)$ y $cos(x)$ en el rango ($-\pi$ , $\pi$) y grafiquen su resultado en un solo panel. 
*Utiliza colores, estilos de línea y marcadores para diferenciar cada función trigonométrica. Además usa todas las funciones necesarias para mejorar tu gráfica.*


````python
angles=np.linspace(-np.pi,np.pi,100)

sen=np.sin(angles)
cos=np.cos(angles)

plt.plot(angles, sen, linestyle="--",label=r"$sen(x)$",marker="*")
plt.plot(angles, cos, label='cos(x)')
# title define el titulo
plt.title('Funciones trigonométricas')
# xlabel define la etiqueta del eje "x"
plt.xlabel('Ángulo ($-\pi$,$\pi$)')
# ylabel define la etiqueta del eje "y"
plt.ylabel('$f (x)=sen(x)$')
# la instruccion legend agrega una leyenda
plt.legend()
# grid con lineas punteadas 
plt.grid(linestyle='--')
````

### <font color="brown"> Gráficos de dos ejes 
Un gráfico de dos ejes permite visualizar dos variables diferentes compartiendo el mismo eje x, pero con escalas distintas en y. Es de gran utilidad cuando tenemos dos variables diferentes con unidades distintas, pero que queremos comparar en el mismo tiempo. Para ello, ocupamos la función `.twinx()`

#### <font color="orange"> Ejemplo: Precipitación vs Temperatura

````python
plt.style.use("default")

meses = np.arange(1, 13)

temperatura = [18,19,21,24,27,30,32,31,29,25,21,19]
precipitacion = [10,15,20,40,80,120,150,140,90,50,20,10]

#Creando una figura y un sistema de ejes donde se dibujará la gráfica
fig, ax1 = plt.subplots()

# Primer eje (izquierdo)
ax1.plot(meses, temperatura, color='#d7263d',label="Temperatura")

#OJO: Las funciones para poner etiquetas cambian
ax1.set_ylabel("Temperatura ($^\circ$C)", color='#d7263d')
ax1.set_xlabel("Mes")
ax1.tick_params(axis='y',labelcolor="red")

# Segundo eje (derecho)
ax2 = ax1.twinx() #Crea un nuevo eje y comparte el mismo eje x    .twiny()
ax2.plot(meses, precipitacion, color='#004dff',label="Precipitación")
ax2.set_ylabel("Precipitación (mm)", color='#004dff')

plt.title("Temperatura [$^\circ$C] vs Precipitación [mm]", color="brown")

#Cuidado cuando usen los grids en este tipo de gráficas
ax1.grid(linestyle="--",alpha=0.3,color="gray")
ax2.grid(linestyle="--",alpha=0.3,color="gray")

#Agregar leyendas
#lines = [line1, line2]
#labels = [line.get_label() for line in lines]
#ax1.legend(lines, labels, loc='upper left')

plt.show()
````

````python
#Agregar leyendas a la siguiente tabla

plt.style.use("default")

meses = np.arange(1, 13)

temperatura = [18,19,21,24,27,30,32,31,29,25,21,19]
precipitacion = [10,15,20,40,80,120,150,140,90,50,20,10]

#Creando una figura y un sistema de ejes donde se dibujará la gráfica
fig, ax1 = plt.subplots()

# Primer eje (izquierdo)
ax1.plot(meses, temperatura, color='#d7263d',label="Temperatura")[0] #Guardo mi plot en una variable

#OJO: Las funciones para poner etiquetas cambian
ax1.set_ylabel("Temperatura ($^\circ$C)", color='#d7263d')
ax1.set_xlabel("Mes")
ax1.tick_params(axis='y',labelcolor="red")

# Segundo eje (derecho)
ax2 = ax1.twinx() #Crea un nuevo eje y comparte el mismo eje x    .twiny()
ax2.plot(meses, precipitacion, color='#004dff',label="Precipitación") #Guardo mi plot en una segunda variable
ax2.set_ylabel("Precipitación (mm)", color='#004dff')

plt.title("Temperatura [$^\circ$C] vs Precipitación [mm]", color="brown")

#Cuidado cuando usen los grids en este tipo de gráficas
ax1.grid(linestyle="--",alpha=0.3,color="gray")
ax2.grid(linestyle="--",alpha=0.3,color="gray")

#Agregar leyenda
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

plt.show()
````

## <font color="darkblue"> Gráficos de dispersión

El gráfico de dispersión (scatter plot), se utilizan usualmente para comparar los valores que toman dos variables distintas. Ambas representadas en diferentes ejes. La gráfica resultante permite identificar visualmente la *posible* correlación entre ambas variables. Para hacer uso de este tipo de gráficos se usa la función `plt.scatter`.

#### <font color="orange">Ejemplo 1: Gráfico de puntos básico

````python
# Generar datos
x = np.array([3, 8, 5, 6, 1, 9, 6, 7, 2, 1, 8])
y = np.array([4, 5, 2, 4, 6, 1, 4, 6, 5, 2, 3]) # Nota: Recuerden que el largo de ambas listas debe de ser igual

# Graficar
plt.figure(figsize=(5,5))
plt.scatter(x = x, y = y)
plt.grid(alpha=0.2)
plt.xlabel("Posición x")
plt.ylabel("Posición y")
plt.show()
````

### <font color="brown"> Personalización de scatter plots

* **Marcador** (`marker`)
* **Color** (`color`o `c`)
* **Tamaño** (`s`)
* **Contorno** (`edgecolor`)
* **Grosor del contorno** (`linewidths`)
* **Transparencia** (`alpha`)

#### <font color="orange"> Ejemplo 2: Scatter plot personalizado

````python
#Definimos valores randoms par x y y
np.random.seed(42)
x = np.random.randn(50)
y = np.random.randn(50)
 
plt.figure(figsize=(6, 4))

plt.scatter(x, y,
    s=100,              # tamaño del marcador
    c='steelblue',      # color
    marker='*',         # forma del marcador
    alpha=0.7,          # transparencia
    edgecolors='black', # color del borde
    linewidths=0.5,     # ancho del borde
)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de Dispersión Personalizado')
plt.show()
````

#### <font color="orange"> Ejemplo 3: Colormaps
También pueden usar colormaps pero para eso deben de escoger los ticks en los que irá cambiando la paleta de color y deben de ser del mismo tamaño que sus datos (<https://matplotlib.org/stable/users/explain/colors/colormaps.html>).

````python
np.random.seed(1)

# Coordenadas simuladas (México)
latitud = np.random.uniform(14, 32, 80) #No queremos sesgo hacia ningún valor
longitud = np.random.uniform(-118, -86, 80)

# Temperatura simulada 
temperatura = np.random.randint(10,35, 80) #Misma longitud que sus datos

plt.scatter(longitud,
            latitud,  
            c=temperatura, #ticks asociados al colorbar
            cmap="inferno", #El cmap escoge una rampa de colores de acuerdo al valor dado
            marker = "o", 
            s=60, 
            alpha = 1, 
            edgecolors = "black", 
            linewidths = 1)

#Definimos la barra de color
plt.colorbar(label="Temperatura (°C)")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.title("Distribución espacial de temperatura")

plt.show()
````

````python
#Segunda forma

x1 = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y1 = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

#Separamos los subconjuntos de datos mediante condiciones
subconjunto = np.where(x1 < 7, "yellow", "lightblue")
print(subconjunto)

plt.scatter(x1, y1, color = subconjunto)

plt.title("Segunda forma")
plt.xlabel("Datos x")
plt.ylabel("Datos y")
#¿Acá cómo agregarían la leyenda?


plt.show()
````

````python
#Tercer manera
x1 = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y1 = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

xsub1=[]
xsub2=[]
ysub1=[]
ysub2=[]

for i,jj in enumerate (x1):
    if jj<7:
        ysub1.append(y1[i])
        xsub1.append(x1[i])
    else:       
        ysub2.append(y1[i])
        xsub2.append(x1[i])

print(ysub1)
print(xsub1)

plt.scatter(xsub1,ysub1,label="sub1")
plt.scatter(xsub2,ysub2,label="sub2")
plt.title ("Tercera forma")
plt.legend()

plt.show()
````

````python
#Parte 2

np.random.seed(90)

X2=np.random.uniform(200,500,100)
Y2=np.random.uniform(400,800,100)

subx1,suby1,subx2,suby2,subx3,suby3=[],[],[],[],[],[]


for i,ii in enumerate (X2):
    if ii<250:
        subx1.append(Y2[i])
        suby1.append(X2[i])
        
    elif 350>ii<400 :
        subx2.append(Y2[i])
        suby2.append(X2[i])

    else:
        
        suby3.append(Y2[i])
        subx3.append(X2[i])

plt.scatter(subx1,suby1,label="Subconjunto 1",s=100,marker="3")
plt.scatter(subx2,suby2,label="Subconjunto 2",s=100,marker="p")
plt.scatter(subx3,suby3,label="Subconjunto 3",s=100,marker="h")

plt.xlabel("X2")
plt.ylabel("Y2")

plt.legend()
plt.title("Ejercicio 2-Parte 2")

plt.show()
````
