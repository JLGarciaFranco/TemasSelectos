### **<font color="DodgerBlue">Ejercicio 1 — Perfil CTD con muestreo irregular: interpolación y contenido de calor oceánico</font>**

<font color="DarkBlue">

Un **CTD** (*Conductivity, Temperature, Depth*) es el instrumento oceanográfico de referencia para caracterizar la estructura vertical de la columna de agua. La sonda mide simultáneamente tres variables: **conductividad** (de la que se deriva la salinidad), **temperatura** y **presión** (convertida a profundidad mediante la ecuación hidrostática). Combinando las tres es posible calcular la densidad del agua de mar con la ecuación de estado del océano, lo que permite identificar estructuras como la termoclina, la haloclina y estudiar la estabilidad vertical de la columna.

La sonda no mide a profundidades fijas: registra a intervalos de tiempo constantes mientras desciende a velocidad aproximadamente uniforme (~1 m/s). El resultado es un perfil con **muestreo en profundidad no uniforme** — los niveles superficiales, donde los gradientes son más intensos, suelen tener mayor resolución que las aguas profundas. Esta característica hace que la interpolación y la integración de perfiles CTD sean más delicadas que con series de resolución regular.

El archivo `ctd_cast_A.csv` está disponible en la **carpeta `data/` del curso**. Contiene dos columnas: `depth_m` (profundidad en metros) y `temp_C` (temperatura en °C), correspondientes a un perfil de estación única en el Pacífico Norte.

Esto es un filtro digital en el dominio de la frecuencia — una herramienta fundamental en el análisis de series geofísicas.
