Los números de Harshad (del sánscrito "alegría" y "dar") son números enteros positivos que son divisibles por la suma de sus dígitos.

Ejemplos:

    12: Sus dígitos suman 1 + 2 = 3, y 12 ÷ 3 = 4 (sin residuo) ✓
    18: Sus dígitos suman 1 + 8 = 9, y 18 ÷ 9 = 2 (sin residuo) ✓
    11: Sus dígitos suman 1 + 1 = 2, pero 11 ÷ 2 = 5.5 (con residuo) ✗

* **Función que grafique la trayectoria de un asteroide**. Para ello, ocupen la ecuación orbital:


   $$
   r(\theta) = \frac{a(1 - e^2)}{1 + e \cos(\theta)}
   $$
  

    donde $a$ = semieje mayor (en UA), $e$= excentricidad (0>$e$<1), $r$ = distancia al Sol, y $\theta$ es el ángulo polar que describe la posición del cuerpo en su órbita y va de 0 a 2$\pi$.

   `Nota`:*La ecuación orbital regresa la posición en coordenadas polares($r$,$\theta$), por lo que para que puedan graficar la trayectoria, será necesario que hagan la conversión a coordenadas cartesianas ($x$,$y$)*:
    $$
    x = r(\theta) * cos(\theta)
    $$

   $$
    y = r(\theta)* seno(\theta)
   $$
