# <font color="purple"> **Ejercicio 3: Carbono negro**

El **carbono negro (BC)**, después del dióxido de carbono, es uno de los contaminantes que contribuyen al calentamiento global. Forma parte de los llamados contaminantes climáticos de vida corta (CCVC), los cuales tienen un importante potencial de calentamiento global y un tiempo de vida en la atmósfera más corto que el $CO_2$. Se produce por diferentes actividades del hombre como la combustión incompleta de combustibles fósiles como el diésel y el combustóleo, así como por la quema de leña, entre otros. El efecto del carbono negro en el ambiente no sólo produce el calentamiento de la atmósfera, también deteriora la calidad del aire y se le ha asociado con efectos negativos graves sobre la salud humana, daños a diversos ecosistemas y a la infraestructura urbana.

En el 2017, la Secretaría de Medio Ambiente y Recursos Naturales publicó un [artículo](<https://www.gob.mx/semarnat/articulos/que-es-el-carbono-negro-134747>) en donde se comentaba que México se había propuesto reducir sus emisiones de BC. La meta establecida contempla la reducción no condicionada del 51% del volumen de sus emisiones para el año 2030 tomando como referencia un escenario tendencial carente de medidas para combatir el cambio climático.
